1. ReLU:"Rectified Linear Unit"의 줄임말로, 정류 선형 유닛이라고 번역됩니다.
o 기본개념: ReLU(x) = max(0, x)
  - 입력값이 0보다 크면 그대로 출력
  - 입력값이 0 이하면 0을 출력
o 동작 방식:
  - 입력이 5라면 → 출력은 5
  - 입력이 -3이라면 → 출력은 0
  - 입력이 0이라면 → 출력은 0
o 주요 장점:
  - 계산이 매우 간단: 단순한 비교 연산만 필요
  - 기울기 소실 문제 완화: 양수 구간에서 기울기가 1로 일정
  - 희소성: 음수 입력을 0으로 만들어 네트워크를 희소하게 만듦
  - 빠른 학습: 계산 복잡도가 낮아 학습 속도가 빠름
o 단점:
  - Dying ReLU 문제 - 뉴런이 음수 영역에 빠지면 영원히 0만 출력할 수 있음
o 사용 위치:
  - 주로 신경망의 은닉층(hidden layer)에서 사용되며, 출력층에서는 문제 유형에 따라 다른 활성화 함수를 사용합니다.
  - ReLU의 등장으로 깊은 신경망 학습이 훨씬 효과적으로 가능해졌고, 현재도 가장 기본적인 활성화 함수로 사용되고 있습니다.

2. Softmax: 머신러닝과 딥러닝에서 자주 사용되는 활성화 함수입니다.
o 기본 개념: Softmax 함수는 여러 개의 실수 값을 입력받아서, 각각을 0과 1 사이의 확률 값으로 변환하고, 모든 출력 값의 합이 1이 되도록 정규화하는 함수
  - 수식: 입력 벡터가 [x₁, x₂, ..., xₙ]일 때, i번째 원소의 softmax 값은, softmax(xᵢ) = eˣⁱ / (eˣ¹ + eˣ² + ... + eˣⁿ)
o 주요 특징:
  - 모든 출력 값이 0과 1 사이의 값을 가집니다
  - 모든 출력 값의 합이 정확히 1입니다
  - 입력 값이 클수록 해당하는 출력 확률이 높아집니다
o 사용 예시:
  - 다중 클래스 분류 문제에서 신경망의 마지막 층에 주로 사용됩니다. 예를 들어 고양이, 개, 새를 구분하는 모델에서 출력이 [2.0, 1.0, 0.1]이라면, softmax를 적용하면 약 [0.66, 0.24, 0.10]이 되어 "고양이일 확률 66%, 개일 확률 24%, 새일 확률 10%"로 해석할 수 있습니다.
o 왜 사용하나요?
  - Raw 점수들을 확률 분포로 변환해서 해석하기 쉽게 만들고, 크로스 엔트로피 손실 함수와 함께 사용하여 효과적인 학습이 가능하기 때문입니다.

3. ReLU와 softmax를 함께 사용하는 이유: 신경망에서 각각이 서로 다른 역할을 담당하며, 상호 보완적으로 작용하기 때문입니다.
o 역할 분담:
  - ReLU: 은닉층(hidden layers)에서 사용
  - Softmax: 출력층(output layer)에서 사용
o 함께 사용하는 핵심 이유:
  1. 단계별 처리
  - 입력 → [은닉층 + ReLU] → [은닉층 + ReLU] → ... → [출력층 + Softmax] → 확률 분포
  - ReLU가 중간 과정에서 복잡한 패턴을 학습
  - Softmax가 최종 결과를 확률로 변환

  2. ReLU의 기여
  - 비선형성 추가: 복잡한 데이터 패턴 학습 가능
  - 효율적 학습: 기울기 소실 문제 완화로 깊은 네트워크 학습 가능
  - 희소 표현: 불필요한 뉴런을 꺼서 더 효율적인 표현 학습
  3. Softmax의 기여
  - 확률 해석: 각 클래스에 대한 확신도를 확률로 표현
  - 안정적 학습: Cross-entropy loss와 결합하여 안정적인 역전파
o 실제 예시 (이미지 분류):
  - 이미지 → [Conv+ReLU] → [Conv+ReLU] → [FC+ReLU] → [FC+Softmax] → [고양이: 0.7, 개: 0.2, 새: 0.1]
o 시너지 효과:
  - ReLU가 복잡한 특징을 효과적으로 추출
  - Softmax가 이 특징들을 해석 가능한 확률로 변환
  - 둘 다 계산이 간단해서 전체 네트워크가 효율적

☞ 결국 ReLU는 "좋은 표현을 학습하는 도구"이고, Softmax는 "그 표현을 확률로 번역하는 도구"로서, 각각의 강점을 살려 전체 시스템의 성능을 극대화하는 것!!

```python
# 🧠 간단한 CNN 모델 생성
def create_simple_cnn():
    """
    교육용 간단한 CNN 모델 생성
    """
    # Sequential 모델 생성 - 레이어를 순차적으로 쌓는 방식
    model = tf.keras.Sequential([
        # 첫 번째 컨볼루션 레이어
        # Conv2D(필터개수, 필터크기, 활성화함수, 입력shape)
        # 16: 16개의 서로 다른 특징 검출기(필터) 사용(엣지검출(수직선/수평선/대각선), 색상변화(밝음<->어둠, 색상경계), 모서리와 코더검출(좌상단/우하단,곡선))
        # (3, 3): 3x3 크기의 필터로 입력이미지 스캔
        # activation='relu': 음수 값을 0으로 만들어 비선형성 추가
        # input_shape=(64, 64, 3): 64x64 픽셀, RGB 3채널 이미지 입력
        tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(64, 64, 3)),

        # MaxPooling2D(풀링크기): 2x2 영역에서 최댓값만 선택
        # 이미지 크기를 절반으로 줄임 (64x64 → 32x32)
        # 연산량 감소 + 중요한 특징만 보존
        tf.keras.layers.MaxPooling2D(2, 2),

        # 두 번째 컨볼루션 레이어
        # 32개 필터로 증가 - 더 많은 특징 패턴 학습
        # 이전 레이어의 특징맵을 입력으로 받아 더 복잡한 패턴 인식
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),

        # 두 번째 풀링: 32x32 → 16x16 크기 축소
        tf.keras.layers.MaxPooling2D(2, 2),

        # 세 번째 컨볼루션 레이어
        # 64개 필터 - 가장 복잡하고 추상적인 특징 추출(개발자가 필터종류 선택)
        # 깊은 레이어일수록 고수준 의미를 가진 특징 학습
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),

        # 세 번째 풀링: 16x16 → 8x8 크기 축소
        tf.keras.layers.MaxPooling2D(2, 2),

        # Flatten & Dense 레이어
        # Flatten(): 차원변환기로 다차원 배열을 1차원으로 펼침
        # (8, 8, 64) → (4096,) 변환하여 Dense 레이어 입력 준비 --> (8 x 8)크기로 64채널: 8 x 8 x 64 = 4096
        tf.keras.layers.Flatten(),

        # Dense(뉴런수, 활성화함수): 완전연결층
        # 128개 뉴런(특징을 찾는 감지기)이 모든 입력(4096개)과 연결
        # 뉴런 1개의 계산 과정: 뉴런_출력 = ReLU(w1×입력1 + w2×입력2 + w3×입력3 + ... + w4096×입력4096 + bias)
        # 뉴런 1: "동물성 특징 감지기"
        #   - 가중치들 = [0.8, 0.9, 0.7, -0.1, -0.2, ...]  # 동물 특징엔 양수, 기계 특징엔 음수
        #   - 뉴런1_출력 = ReLU(0.8×털값 + 0.9×눈값 + 0.7×귀값 + (-0.1)×바퀴값 + (-0.2)×금속값 + ...)
        # 뉴런 2: "기계성 특징 감지기"
        #   - 가중치들 = [-0.1, -0.2, -0.1, 0.9, 0.8, ...]  # 동물 특징엔 음수, 기계 특징엔 양수
        #   - 뉴런2_출력 = ReLU(-0.1×털값 + (-0.2)×눈값 + (-0.1)×귀값 + 0.9×바퀴값 + 0.8×금속값 + ...)
        # 파라미터 수 계산: Dense(128): 4096 × 128 + 128 = 524,416개 파라미터 --> 뉴런이 많을수록 정확해지나 복잡성 증가(보통 125~256개)!!
        # 훈련 과정에서 뉴런들이 서로 다른 역할 찾기(훈련과정에서 반응이 있는 파라미터를 랜덤으로 찾아 특징점 파라미터(뉴런)를 정함)
        # 훈련 1회차:
        # 뉴런1: "털 특징에 조금 더 반응하네?" → 털 관련 가중치 조금 증가
        # 뉴런2: "바퀴 특징에 조금 더 반응하네?" → 바퀴 관련 가중치 조금 증가
        #
        # 훈련 100회차:
        # 뉴런1: "털 특징 전문가로 발전 중..."
        # 뉴런2: "바퀴 특징 전문가로 발전 중..."
        #
        # 훈련 1000회차:
        # 뉴런1: "털 패턴 전문가 완성!"
        # 뉴런2: "바퀴 모양 전문가 완성!"
        tf.keras.layers.Dense(128, activation='relu'),

        # Dropout(비율): 과적합 방지 기법 --> 훈련 중 일부 뉴런을 랜덤하게 끄면서 서로다른 패턴으로 학습
        # 훈련 시 50%의 뉴런을 임의로 비활성화
        # 특정 뉴런에 과도하게 의존하는 것을 방지
        # 예측시에는 모든 뉴런사용
        tf.keras.layers.Dropout(0.5),

        # 출력층: 128개의 뉴런을 통해 최종 새로운 3개 뉴런을 생성(예. 자동차/사람/고양이)
        # softmax: 3개 값의 합이 1이 되도록 정규화 (확률 분포)
        tf.keras.layers.Dense(3, activation='softmax')  # 3 classes: Animal/Car/Other
    ])

    # 모델 컴파일 - 학습에 필요한 설정 정의
    model.compile(
        # optimizer: 가중치 업데이트 방법 (Adam은 학습률 자동 조절)
        optimizer='adam',
        # loss: 손실함수 (다중클래스 분류용)
        loss='categorical_crossentropy',
        # metrics: 학습 과정에서 모니터링할 지표
        metrics=['accuracy']
    )

    print("🧠 CNN 모델 생성 완료!")
    return model

# 📊 CNN 구조 시각화
def visualize_model_architecture(model):
    """
    CNN 모델 구조를 시각적으로 보여주기
    \n: 줄바
    "=" * 50: ====...........===== 50개
    """
    print("\n📋 CNN 모델 구조:")
    print("=" * 50)
    # model.summary(): 모델의 레이어별 정보 출력
    # 각 레이어의 출력 크기와 파라미터 수 확인 가능
    model.summary()

    # 레이어별 설명
    print("\n🔍 레이어별 역할:")
    print("📌 Conv2D: 특징 추출 (엣지, 패턴 등)")
    print("📌 MaxPooling2D: 크기 축소 + 중요 특징 선택")
    print("📌 Flatten: 2D → 1D 변환")
    print("📌 Dense: 최종 분류 결정")
    print("📌 Dropout: 과적합 방지")

# 🎲 가짜 데이터로 빠른 훈련
def quick_train_with_dummy_data(model):
    """
    데모용 가짜 데이터로 빠른 훈련
    """
    print("\n🎓 데모용 빠른 훈련 시작...")

    # 가짜 훈련 데이터 생성 (200개 샘플)
    # np.random.rand(): 0~1 사이 균등분포 난수 생성
    # (200, 64, 64, 3): 200개 샘플, 64x64 픽셀, RGB 3채널
    # .astype('float32'): 메모리 효율성을 위해 32비트 실수형으로 변환
    X_train = np.random.rand(200, 64, 64, 3).astype('float32')

    # 가짜 라벨 생성
    # np.random.randint(0, 3, 200): 0,1,2 중 랜덤하게 200개 선택
    # to_categorical(): 정수 라벨을 원-핫 벡터(분류를 다차원 벡터로 표현)로 변환(예: 1 → [0, 1, 0])
    y_train = tf.keras.utils.to_categorical(np.random.randint(0, 3, 200), 3)

    # 빠른 훈련 (3 epochs만)
    # epochs=3: 전체 (훈련)데이터를 3번 반복 학습
    # batch_size=32: 한 번에 32개 샘플씩 처리 --> 전체 데이터를 나누어 학습하면, 매 회차별로 가중치 업데이트 가능(학습효과 증대)!!
    # verbose=1: 훈련 진행상황을 상세히 출력
    history = model.fit(
        X_train, y_train,
        epochs=3,
        batch_size=32,
        verbose=1
    )

    # history: 모델 훈련 과정의 모든 통계와 기록
    print("✅ 훈련 완료! (실제 프로젝트에서는 실제 데이터 사용)")
    return history

# 🔍 CNN 필터 시각화
def visualize_cnn_filters(model):
    """
    CNN 첫 번째 레이어의 학습된 필터들 시각화
    """
    try:
        # 모델이 빌드되었는지 확인
        # hasattr(): 객체가 특정 속성을 가지고 있는지 확인
        if not hasattr(model, 'built') or not model.built:
            print("⚠️ 모델을 빌드하는 중...")
            # np.random.rand(이미지갯수, 높이, 너비, 색상채널): 0과 1 사이의 랜덤한 실수 생성
            # _: 결과값을 받지만 사용하지 않겠다는 의미
            dummy_input = np.random.rand(1, 64, 64, 3)
            _ = model(dummy_input)

        # 첫 번째 Conv2D 레이어 찾기
        first_conv_layer = None
        # model.layers: 모델의 모든 레이어를 순서대로 반환
        for layer in model.layers:
            # isinstance(): 객체가 특정 클래스의 인스턴스(클래스로부터 만들어진 실제 객체)인지 확인
            if isinstance(layer, tf.keras.layers.Conv2D):
                first_conv_layer = layer
                break  # 첫 번째 Conv2D 레이어만 선택

        if first_conv_layer is None:
            print("❌ Conv2D 레이어를 찾을 수 없습니다.")
            return

        # 첫 번째 Conv2D 레이어의 가중치 추출
        # get_weights(): [필터가중치, 편향] 리스트 반환
        weights = first_conv_layer.get_weights()
        if len(weights) == 0:
            print("⚠️ 아직 가중치가 초기화되지 않았습니다.")
            return

        # weights[0]: 필터 가중치 (shape: [height, width, input_channels, output_channels])
        filters = weights[0]  # 필터 가중치

        print(f"\n🔍 첫 번째 레이어 필터 시각화")
        print(f"필터 개수: {filters.shape[3]}개")  # 출력 채널 = 필터 개수
        print(f"필터 크기: {filters.shape[0]}x{filters.shape[1]}")  # 높이 x 너비

        # 필터 중 처음 8개만 시각화
        # min(): 더 작은 값 선택 (필터가 8개 미만일 수도 있음) --> 화면 공간 제약, 결과의 가독성 향상, 화면 크기 최적화
        num_filters_to_show = min(8, filters.shape[3])

        # plt.subplots(): 여러 서브플롯 생성
        # 2행 4열 배치, figsize로 전체 크기 설정
        fig, axes = plt.subplots(2, 4, figsize=(12, 6))
        fig.suptitle('CNN Learned Filters (First Layer)', fontsize=16)

        for i in range(num_filters_to_show):
            # 서브플롯 위치 계산: i//4 (행), i%4 (열)
            ax = axes[i // 4, i % 4]

            # 필터를 시각화하기 위해 정규화
            # [:, :, 0, i]: 높이, 너비, 첫번째 입력채널, i번째 필터
            filter_img = filters[:, :, 0, i]  # 첫 번째 채널의 i번째 필터

            # 정규화 (0-1 범위로)
            # 최댓값과 최솟값의 차이로 나누어 정규화
            if filter_img.max() > filter_img.min():
                filter_img = (filter_img - filter_img.min()) / (filter_img.max() - filter_img.min())

            # imshow(): 2D 배열을 이미지로 시각화
            # cmap='viridis': 색상맵 설정 (파란색-초록색-노란색)
            ax.imshow(filter_img, cmap='viridis')
            ax.set_title(f'Filter {i+1}')
            ax.axis('off')  # 축 눈금 숨기기

        # tight_layout(): 서브플롯 간격 자동 조정
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"⚠️ 필터 시각화 중 오류 발생: {str(e)}")
        print("💡 이는 모델 구조나 가중치 이슈일 수 있습니다.")
        print("📝 주요 기능(예측)은 정상 작동합니다!")

# 📸 이미지 업로드 및 전처리
def upload_and_preprocess_image():
    """
    이미지 업로드 및 CNN 입력용 전처리
    """
    print("📸 이미지를 업로드해주세요!")
    # files.upload(): Google Colab에서 로컬 파일 업로드
    # 딕셔너리 형태로 {파일명: 파일데이터} 반환
    uploaded = files.upload()

    # 업로드된 파일 정보 추출
    # list().keys()[0]: 딕셔너리의 첫 번째 키(파일명) 선택
    filename = list(uploaded.keys())[0]
    image_data = uploaded[filename]

    # 이미지 로드 및 전처리
    # io.BytesIO(): 바이트 데이터를 파일 객체처럼 처리
    # Image.open(): PIL로 이미지 열기
    image = Image.open(io.BytesIO(image_data))

    # RGB로 변환 (RGBA인 경우)
    # image.mode: 이미지의 색상 모드 확인
    if image.mode != 'RGB':
        image = image.convert('RGB')  # RGB 모드로 강제 변환

    # 크기 조정 (64x64)
    # resize(): 이미지 크기 변경, CNN 입력 크기에 맞춤
    image_resized = image.resize((64, 64))

    # 배열로 변환 및 정규화
    # np.array(): PIL Image를 NumPy 배열로 변환
    # .astype('float32'): 계산 효율성을 위해 32비트 실수형으로 변환
    image_array = np.array(image_resized).astype('float32') / 255.0

    # 배치 차원 추가 (1, 64, 64, 3)
    # np.expand_dims(): 지정된 축에 새로운 차원 추가
    # axis=0: 첫 번째 축에 배치 차원 추가
    # (64, 64, 3) → (1, 64, 64, 3)
    image_batch = np.expand_dims(image_array, axis=0)

    return image, image_resized, image_batch, filename

# 🎯 CNN 예측 및 결과 시각화
def predict_and_visualize(model, original_img, processed_img, image_batch, filename):
    """
    CNN으로 예측하고 결과 시각화
    """
    # 클래스 라벨 정의
    class_names = ['Animal', 'Car', 'Other']

    # CNN 예측
    # model.predict(): 모델로 예측 수행
    # verbose=0: 예측 진행상황 출력 안함
    # 결과: (1, 3) 크기의 확률 배열
    predictions = model.predict(image_batch, verbose=0)

    # 가장 높은 확률의 클래스 선택
    # np.argmax(): 최댓값의 인덱스 반환
    # predictions[0]: 첫 번째 배치의 예측 결과
    predicted_class = np.argmax(predictions[0])

    # 예측된 클래스의 확률값
    confidence = predictions[0][predicted_class]

    # 결과 시각화
    # plt.figure(): 새로운 그림 생성
    plt.figure(figsize=(15, 5))

    # 원본 이미지
    # plt.subplot(행, 열, 인덱스): 서브플롯 위치 지정
    plt.subplot(1, 3, 1)
    plt.imshow(original_img)  # 이미지 표시
    plt.title(f'original image\n({filename})')
    plt.axis('off')  # 축 정보 숨기기

    # 전처리된 이미지 (CNN 입력)
    plt.subplot(1, 3, 2)
    plt.imshow(processed_img)
    plt.title('CNN input image\n(64x64 크기 조정)')
    plt.axis('off')

    # 예측 결과
    plt.subplot(1, 3, 3)
    # plt.bar(): 막대 그래프 생성
    bars = plt.bar(class_names, predictions[0])
    # 가장 높은 확률의 막대를 빨간색으로 강조
    bars[predicted_class].set_color('red')  # 최고 확률 클래스 강조
    plt.title(f'CNN Prediction Results\nPrediction: {class_names[predicted_class]} ({confidence:.2%})')
    plt.ylabel('Probability')
    plt.ylim(0, 1)  # y축 범위 설정

    # 확률 값 표시
    # zip(): 두 리스트를 쌍으로 묶어서 반복
    # enumerate(): 인덱스와 값을 함께 반환
    for i, (name, prob) in enumerate(zip(class_names, predictions[0])):
        # plt.text(): 텍스트 추가 (x좌표, y좌표, 텍스트, 정렬방식)
        plt.text(i, prob + 0.02, f'{prob:.2%}', ha='center')

    plt.tight_layout()
    plt.show()

    # 결과 출력
    print("\n🎯 CNN Prediction Results:")
    print("=" * 30)
    for i, (name, prob) in enumerate(zip(class_names, predictions[0])):
        # 삼항 연산자: 조건 ? 참일때값 : 거짓일때값
        marker = "👉" if i == predicted_class else "  "
        print(f"{marker} {name}: {prob:.2%}")
    print("=" * 30)
    print(f"Final Prediction: {class_names[predicted_class]} (Confidence: {confidence:.2%})")

# 🔬 CNN 중간 레이어 활성화 시각화
def visualize_intermediate_activations(model, image_batch):
    """
    CNN 중간 레이어들의 활성화 맵 시각화
    """
    print("\n🔬 CNN 내부 작동 과정 시각화...")

    try:
        # 모델이 빌드되었는지 확인
        if not hasattr(model, 'built') or not model.built:
            print("⚠️ 모델을 빌드하는 중...")
            # 더미 데이터로 모델 빌드
            dummy_input = np.random.rand(1, 64, 64, 3)
            _ = model(dummy_input)

        # Conv2D 레이어만 찾기
        conv_layers = []
        layer_names = []

        # enumerate(): 인덱스와 값을 함께 반환
        for i, layer in enumerate(model.layers):
            if isinstance(layer, tf.keras.layers.Conv2D):
                conv_layers.append(layer)
                # len(): 리스트 길이 = 현재까지 찾은 Conv2D 레이어 수
                layer_names.append(f'Conv2D Layer {len(conv_layers)}')

        if len(conv_layers) == 0:
            print("❌ Conv2D 레이어를 찾을 수 없습니다.")
            return

        # 중간 레이어 출력을 위한 모델 생성
        # 리스트 컴프리헨션: [expression for item in iterable]
        # 각 Conv2D 레이어의 출력을 수집
        layer_outputs = [layer.output for layer in conv_layers]

        # 새로운 모델 생성: 원본 입력 → Conv 레이어들의 출력
        # inputs=model.input: 원본 모델과 같은 입력 사용
        # outputs=layer_outputs: Conv 레이어들의 출력을 새 모델의 출력으로 설정
        activation_model = tf.keras.models.Model(inputs=model.input, outputs=layer_outputs)

        # 활성화 맵 계산
        # 실제 이미지를 중간 레이어 출력 모델에 통과시킴
        activations = activation_model.predict(image_batch, verbose=0)

        # 단일 출력인 경우 리스트로 변환
        if not isinstance(activations, list):
            activations = [activations]

        # 시각화
        # min(): 최대 3개 레이어만 표시 (화면 공간 제약)
        num_layers = min(3, len(conv_layers))
        plt.figure(figsize=(15, 10))

        for i in range(num_layers):
            activation = activations[i]  # i번째 레이어의 활성화 맵
            layer_name = layer_names[i]

            # 처음 4개 필터만 표시
            # activation.shape[-1]: 마지막 차원 = 필터(채널) 수
            num_filters = min(4, activation.shape[-1])
            for j in range(num_filters):
                # 서브플롯 위치 계산: (레이어번호*4 + 필터번호 + 1)
                plt.subplot(num_layers, 4, i*4 + j + 1)

                # 활성화 맵 정규화
                # activation[0, :, :, j]: 첫 번째 배치, 모든 높이, 모든 너비, j번째 필터
                feature_map = activation[0, :, :, j]

                # 정규화 처리
                if feature_map.max() > feature_map.min():
                    feature_map = (feature_map - feature_map.min()) / (feature_map.max() - feature_map.min())

                plt.imshow(feature_map, cmap='viridis')
                plt.title(f'{layer_name}\nFilter {j+1}')
                plt.axis('off')

        # plt.suptitle(): 전체 그림의 제목 설정
        plt.suptitle('CNN Feature Maps - How CNN "Sees" Your Image', fontsize=16)
        plt.tight_layout()
        plt.show()

        print("💡 해석:")
        print("- 첫 번째 레이어: 기본적인 엣지, 색상 검출")
        print("- 두 번째 레이어: 더 복잡한 패턴 조합")
        print("- 세 번째 레이어: 고수준 특징 (객체 부분)")

    except Exception as e:
        print(f"⚠️ 활성화 시각화 중 오류 발생: {str(e)}")
        print("💡 이는 모델 구조나 TensorFlow 버전 이슈일 수 있습니다.")
        print("📝 주요 기능(예측)은 정상 작동합니다!")

# 🎮 메인 실행 함수
def run_cnn_demo():
    """
    CNN 교육용 데모 메인 실행
    """
    print("🎉 진짜 CNN 교육용 데모 시작!")
    print("=" * 50)

    # 1. CNN 모델 생성
    # 함수 호출하여 Sequential 모델 객체 생성
    model = create_simple_cnn()

    # 2. 모델 구조 확인
    # 생성된 모델의 레이어 정보 출력
    visualize_model_architecture(model)

    # 3. 빠른 훈련 (데모용)
    # 가짜 데이터로 모델 학습 진행
    history = quick_train_with_dummy_data(model)

    # 4. 학습된 필터 시각화
    # 첫 번째 Conv2D 레이어의 필터들 시각화
    visualize_cnn_filters(model)

    # 5. 이미지 업로드 및 예측
    # 함수 호출하여 4개 값을 동시에 받음 (튜플 언패킹)
    original_img, processed_img, image_batch, filename = upload_and_preprocess_image()

    # 6. CNN 예측 및 결과 시각화
    # 전처리된 이미지로 예측 수행 및 결과 시각화
    predict_and_visualize(model, original_img, processed_img, image_batch, filename)

    # 7. CNN 내부 작동 과정 시각화
    # 중간 레이어들의 활성화 맵 시각화
    visualize_intermediate_activations(model, image_batch)

    print("\n🎓 CNN 데모 완료!")
    print("💡 이제 CNN이 어떻게 이미지를 '이해'하는지 보셨습니다!")

# 🚀 데모 실행
print("📚 CNN 교육용 데모 - 실제 신경망으로 이미지 분류 체험")
print("🔥 이번에는 진짜 CNN입니다!")
print()
# 메인 함수 호출하여 전체 데모 실행
run_cnn_demo()
```
